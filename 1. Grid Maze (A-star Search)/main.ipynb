# Comparative statistics
In this section we will see some comparative statistics. In particular, increasing the number of obstacles, we will see how the different combinations of algorithms and heuristics will behave in order to get the path to the goal.

I set a maximum number of obstacles of 100 and for each number of obstacles each combination of heuristic and algorithm will iterate 100 times. 

The increase in the number of obstacles is halted when, for a certain value, all 100 iterations fail to find a path to the goal. This indicates that the grid has become saturated with obstacles, making it pointless to continue adding more.

The statistics function is designed to be scalable, allowing for easy integration of additional heuristics or algorithms.


```python
heuristics = [
    h_manhattan,
    h_manhattan_obstacle_weighted,
    h_euclidean
]

algorithms = [
    "a_star",
    "best_first"
]
```
